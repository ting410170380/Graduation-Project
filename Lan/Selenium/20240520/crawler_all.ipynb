```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
 
 
url = 'https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR3QwTlRFU0JYcG9MVlJYS0FBUAE?hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant'
r = requests.get(url)
web_content = r.text
soup = BeautifulSoup(web_content,'lxml')


title = soup.find_all('div', class_='m5k28')
title1 = soup.find_all('div', class_='IL9Cne')
print(title1)
```

```python
titles = [t.find('a').text for t in title1]


print(titles)
```

```python
newUrls = [requests.get(t.find('a')['href'].replace('.','https://news.google.com',1)).url for t in title1]
print(newUrls)
```

```python
df = pd.DataFrame(
{
    'title': titles,
    'links': newUrls
})
 
print(df)
```

```python
error = "如果您看到這個頁面，表示您的網路已遭到停止訪問本網站的權利。"
```

```python
Titles = []
NewUrls = []
ArticleContext = []
#01234
for i in range(len(titles)):
    #print(i)
    url = df['links'][i]
    #print(url)  #網址
    r = requests.get(url)
    web_content = r.text
    soup = BeautifulSoup(web_content,'lxml')
    articleContent = soup.find_all('p')
    #print(articleContent)
    article = []
    if(articleContent != []):
        if(articleContent[0].text not in error):
            for p in articleContent:
                article.append(p.text)
            articleAll = '\n'.join(article)
            ArticleContext.append(articleAll)
            Titles.append(df['title'][i])
            NewUrls.append(df['links'][i])
        else:
            print("no:", i)
            print(url)
#print(ArticleContext)
    
```

```python
df1 = pd.DataFrame(
{
    'title': Titles,
    'links': NewUrls,
    'context': ArticleContext
})
 
print(df1)
```
