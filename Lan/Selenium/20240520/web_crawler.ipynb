```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
 
 
url = 'https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR3QwTlRFU0JYcG9MVlJYS0FBUAE?hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant'
r = requests.get(url)
web_content = r.text
soup = BeautifulSoup(web_content,'lxml')


title = soup.find_all('div', class_='m5k28')
title1 = soup.find_all('div', class_='IL9Cne')
print(title1)
```

```python
titles = [t.find('a').text for t in title1]


print(titles)
```

```python
newUrls = [requests.get(t.find('a')['href'].replace('.','https://news.google.com',1)).url for t in title1]
print(newUrls)
```

```python
df = pd.DataFrame(
{
    'title': titles,
    'links': newUrls
})
 
print(df)
```

```python
url = df['links'][2]
print(url)
r = requests.get(url)
web_content = r.text
soup = BeautifulSoup(web_content,'lxml')
```

```python
articleContent = soup.find_all('p')
print(articleContent)
```

```python
article = [] 
for p in articleContent:
    article.append(p.text)
```

```python
articleAll = '\n'.join(article)
print(articleAll)
```
